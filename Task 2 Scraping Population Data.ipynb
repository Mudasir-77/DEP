Web Scrapping on Poplulation Data


```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
```

```python
import requests
```

```python
response = requests.get(url)
```

```python
response
```

```python
html_txt = response.text # html format
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html_txt , "html.parser")
```

```python
type(soup)
```

```python
table = soup.find("table", {"class":"wikitable"})
```

```python
table
```

```python

rows = table.find_all("tr")
```

```python
rows # th and td
```

```python
for row in rows:
    th_lst = row.find_all("th")
    for h in th_lst:
        print(h.text.strip())
```

```python
headers = []

for row in rows:
    th_lst = row.find_all("th")
    for h in th_lst:
        headers.append(h.text.strip())
```

```python
print(headers)
```

```python
headers[0]
```

```python
headers[0] = "Rank"
```

```python
headers
```

```python
headers
```

```python
for row in rows:
    td = row.find_all("td")
    for d in td:
        print(d.text.strip())
```

```python

for row in rows:
    td = row.find_all("td")
    table_data = []
    for d in td:
        table_data.append(d.text.strip())
    print(table_data)
```

```python
final_data = []
for row in rows:
    td = row.find_all("td")
    table_data = []
    for d in td:
        table_data.append(d.text.strip())
    final_data.append(table_data)
```

```python
final_data
```

```python
final_data = final_data[1:]
```

```python
final_data[1]
```

```python
final_data[1][0] = "1"
```

```python
final_data[2]
```

```python
final_data[2].insert(0,"2")
```
